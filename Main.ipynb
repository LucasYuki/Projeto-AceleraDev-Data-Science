&emsp;O loss em si não é o parâmetro mais importante na escolha da rede, o modelo escolhido deve ser o que tenha a menor diferença entre o loss do treino e o loss de teste. Isto pois o loss é a média da distância quadratica do centro e a saída da rede no conjunto de dados utilizados, então valores de losses próximos tanto para o conjunto de treino quanto para o de teste significa que esses dados tem comportamentos semelhantes. Para isso o erro foi calculado por $\frac{|Loss-Loss_{Train}|}{Loss_{Train}}$ e o modelo escolhido é o que apresenta menor erro no teste.

```python
losses = pd.read_csv("Deep_One-Class/portfolio_"+str(portifolio)+"/losses.csv", index_col=0)
escolhido = losses["error test"].idxmin()
print("o modelo com menor error test é o modelo", escolhido)
print("com erro de", min(losses["error test"]))
losses
```

&emsp;Com o modelo definido basta aplicá-lo ao resto do conjunto de dados, quanto menor a distância quadratica entre o centro e o valor obtido mais essa empresa se assemelha as empresas já presentes no portifólio do usuário.

```python
Aply_Deep_One(data, "Deep_One-Class/portfolio_"+str(portifolio), portifolio, escolhido)
```
